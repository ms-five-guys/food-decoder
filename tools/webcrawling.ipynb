### ğŸ–±ï¸ì›¹ë“œë¼ì´ë¸Œ ì„¤ì • ë° ì›¹í˜ì´ì§€ ì ‘ê·¼
###### Seleniumì„ ì‚¬ìš©í•˜ì—¬ ì›¹í˜ì´ì§€ì— ì ‘ê·¼í•˜ê³ , BeautifulSoupì„ í†µí•´ HTMLì„ íŒŒì‹±

```python
from selenium import webdriver # Selenium ì›¹ë“œë¼ì´ë²„
from selenium.webdriver.chrome.service import Service # ì›¹ë“œë¼ì´ë²„ ì„œë¹„ìŠ¤ ê´€ë¦¬
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup HTML íŒŒì‹±ì„ ìœ„í•œ BeautifulSoup
import time # í˜ì´ì§€ ë¡œë”© ëŒ€ê¸° ì‹œê°„ ì„¤ì •

# ì›¹ë“œë¼ì´ë²„ ì„¤ì •
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # ë¸Œë¼ìš°ì € ì°½ì„ ë„ìš°ì§€ ì•ŠìŒ
options.add_argument('--disable-dev-shm-usage') # ë©”ëª¨ë¦¬ ì‚¬ìš© ìµœì í™”

# ì›¹ë“œë¼ì´ë²„ ì‹¤í–‰
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# ëŒ€ìƒ ì›¹í˜ì´ì§€ URL
url = 'https://www.allrecipes.com/'

# ì›¹í˜ì´ì§€ ì—´ê¸°
driver.get(url)

# í˜ì´ì§€ ë¡œë”© ëŒ€ê¸°
time.sleep(3)  # í•„ìš”ì— ë”°ë¼ ì¡°ì •

# í˜ì´ì§€ ì†ŒìŠ¤ ê°€ì ¸ì˜¤ê¸°
page_source = driver.page_source

# BeautifulSoupìœ¼ë¡œ HTML íŒŒì‹±
soup = BeautifulSoup(page_source, 'html.parser')

# ì›¹ë“œë¼ì´ë²„ ì¢…ë£Œ
driver.quit()
```

```python
# ì›¹ì—ì„œ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¤ê¸° ìœ„í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬
import requests
 # íŒŒì¼ ë° ë””ë ‰í† ë¦¬ ê´€ë ¨ ì‘ì—…ì„ ìœ„í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬
import os

# ì €ì¥í•  ë””ë ‰í† ë¦¬ ì„¤ì •
save_dir = 'downloaded_images'

# í´ë”ê°€ ì—†ìœ¼ë©´ ìƒì„±
os.makedirs(save_dir, exist_ok=True)

# ì´ë¯¸ì§€ íƒœê·¸ ì°¾ê¸°
# soup.find_all('img'): HTMLì—ì„œ <img> íƒœê·¸ë¥¼ ëª¨ë‘ ì°¾ì•„ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜
images = soup.find_all('img')

# ì´ë¯¸ì§€ ë‹¤ìš´ë¡œë“œ
for idx, img in enumerate(images):
    img_url = img.get('src')
    if img_url:
        try:
            # ì´ë¯¸ì§€ ë°ì´í„° ê°€ì ¸ì˜¤ê¸°(img_urlì—ì„œ HTTP GET ìš”ì²­ì„ ë³´ë‚´ì„œ ì´ë¯¸ì§€ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜´)
            img_data = requests.get(img_url).content
            # íŒŒì¼ ê²½ë¡œ ì„¤ì •
            img_filename = os.path.join(save_dir, f'image_{idx + 1}.jpg')
            # ì´ë¯¸ì§€ ì €ì¥
            with open(img_filename, 'wb') as f:
                f.write(img_data)
            print(f'Successfully downloaded {img_filename}')
        except Exception as e:
            print(f'Failed to download image at {img_url}: {e}')
```
